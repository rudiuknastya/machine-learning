```python
import matplotlib.pyplot as plt
import csv
import numpy as np
x = []
y = []
  
with open('lab_1_train.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    next(lines)
    for column in lines:
        x.append(float(column[1]))
        y.append(float(column[2]))
  
plt.scatter(x, y, color = 'g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Train data', fontsize = 20)
plt.show()
```

```python
alfa = 0.01
tolerance = 0.000001
b0 = 1
b1 = 1
def linear_regresion(b0, b1, x):
    return [b0 + b1*i for i in x]

def cost_function(y_pr, y):
    diff = np.subtract(y, y_pr)
    diff = np.square(diff)
    return np.sum(diff)/len(diff)

y_pred = linear_regresion(b0, b1, x)
cost = cost_function(y_pred, y)
differ = cost

while differ > tolerance:
    y_diff = np.subtract(y, y_pred)
    xy_diff = np.multiply(x,y_diff)
    gradient_b0 = -2*np.sum(y_diff)/len(y_diff)
    gradient_b1 = -2*np.sum(xy_diff)/len(xy_diff)
    b0 = b0 - alfa*gradient_b0
    b1 = b1 - alfa*gradient_b1
    print('b0 =',b0,'b1 =',b1, 'cost =', cost, sep=' ')
    y_pred = linear_regresion(b0, b1, x)
    differ = cost - cost_function(y_pred, y)
    cost = cost_function(y_pred, y)
```

```python
x_test = []
y_test = []
with open('lab_1_test.csv','r') as csvfile:
    line = csv.reader(csvfile, delimiter=',')
    next(line)
    for columns in line:
        x_test.append(float(columns[1]))
        y_test.append(float(columns[2]))

y_testpred = linear_regresion(b0, b1, x_test)
cost_test = cost_function(y_testpred, y_test)
cost_test
```

```python
plt.title('Linear Regression', fontsize = 20)
plt.plot(x, y, 'go',
         x_test, y_test, 'yo',
          [b0+b1*0, b0+b1*1]
         )
plt.show
```
