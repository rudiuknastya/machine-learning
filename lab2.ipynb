```python
from sklearn. preprocessing import OneHotEncoder
encoder = OneHotEncoder()
train_encoder = pd.DataFrame(encoder.fit_transform(trainData[categorical_features]).toarray())
train_encoder
```

```python
newTrain = trainData.join (train_encoder)
newTrain.drop(categorical_features, axis= 1 , inplace= True )
```

```python
newTrain = newTrain.drop("TARGET", axis=1)
```

```python
newTrain
```

```python
test_encoder = pd.DataFrame(encoder.transform(testData[categorical_features]).toarray())
test_encoder
```

```python
newTest = testData.join (test_encoder)
newTest.drop(categorical_features, axis= 1 , inplace= True )
newTest
```

```python
newTrain.columns = newTrain.columns.astype(str)
newTest.columns = newTest.columns.astype(str)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

train_x, test_x, train_y, test_y = train_test_split(newTrain, trainData["TARGET"], test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_train_x = sc.fit_transform(train_x)
scaled_train_x
```

```python
scaled_test_x = sc.transform(test_x)
```

```python
lr=LogisticRegression(max_iter=2000)
lr.fit(scaled_train_x, train_y)
```

```python
train_preds = lr.predict(scaled_train_x)
print("ROC_AUC: ", roc_auc_score(train_y, train_preds))
```

```python
test_preds = lr.predict(scaled_test_x)
print("Test ROC_AUC: ", roc_auc_score(test_y, test_preds))
```

```python
submTest = lr.predict(newTest)
```
